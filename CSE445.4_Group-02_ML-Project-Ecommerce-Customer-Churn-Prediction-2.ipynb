__Here, we can see that, the columns - Tenure, WarehouseToHome, HourSpendOnApp, OrderAmountHikeFromlastYear, CouponUsed, OrderCount, DaySinceLastOrder have NULL values. We have to replace it with the MEDIAN value of respective column.__   

```python
df[['Tenure', 'WarehouseToHome','HourSpendOnApp','OrderAmountHikeFromlastYear','CouponUsed','OrderCount','DaySinceLastOrder']] = df[['Tenure', 'WarehouseToHome','HourSpendOnApp','OrderAmountHikeFromlastYear','CouponUsed','OrderCount','DaySinceLastOrder']].fillna(df[['Tenure', 'WarehouseToHome','HourSpendOnApp','OrderAmountHikeFromlastYear','CouponUsed','OrderCount','DaySinceLastOrder']].median())


df[['PreferredLoginDevice','PreferredPaymentMode','NumberOfDeviceRegistered','MaritalStatus','NumberOfAddress','Complain','CashbackAmount']] = df[['PreferredLoginDevice','PreferredPaymentMode','NumberOfDeviceRegistered','MaritalStatus','NumberOfAddress','Complain','CashbackAmount']].fillna(df[['PreferredLoginDevice','PreferredPaymentMode','NumberOfDeviceRegistered','MaritalStatus','NumberOfAddress','Complain','CashbackAmount']].apply(lambda x: x.fillna(x.value_counts().index[0])))
```

**Data Balancing Technique**

```python
# Churn count
class_count_0, class_count_1 = df['Churn'].value_counts()

# Separate class
class_0 = df[df['Churn'] == 0]
class_1 = df[df['Churn'] == 1]# print the shape of the class
print('class 0:', class_0.shape)
print('class 1:', class_1.shape)
```

```python
# Gender count
class_count_1, class_count_2 = df['Gender'].value_counts()

# Separate class
class_1 = df[df['Gender'] == 'Male']
class_2 = df[df['Gender'] == 'Female']# print the shape of the class
print('class 1:', class_1.shape)
print('class 2:', class_2.shape)
```

```python
class_1_over = class_1.sample(class_count_0, replace=True)

test_over = pd.concat([class_1_over, class_0], axis=0)

print("total class of 1 and 0:\n",test_over['Churn'].value_counts())# plot the count after under-sampeling
test_over['Churn'].value_counts().plot(kind='bar', title='count(Not Churn/Churn)', figsize=(8, 6))
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
X = final_processed_data[['Tenure', 'WarehouseToHome', 'HourSpendOnApp',
       'OrderAmountHikeFromlastYear', 'CouponUsed', 'OrderCount',
       'DaySinceLastOrder', 'CashbackAmount',
       'CityTier_1', 'CityTier_2', 'CityTier_3', 'PreferredPaymentMode_CC',
       'PreferredPaymentMode_COD', 'PreferredPaymentMode_Cash on Delivery',
       'PreferredPaymentMode_Credit Card', 'PreferredPaymentMode_Debit Card',
       'PreferredPaymentMode_E wallet', 'PreferredPaymentMode_UPI',
       'Gender_Female', 'Gender_Male', 'PreferedOrderCat_Fashion',
       'PreferedOrderCat_Grocery', 'PreferedOrderCat_Laptop & Accessory',
       'PreferedOrderCat_Mobile', 'PreferedOrderCat_Mobile Phone',
       'PreferedOrderCat_Mobiles', 'PreferedOrderCat_Others',
       'SatisfactionScore_1', 'SatisfactionScore_2', 'SatisfactionScore_3',
       'SatisfactionScore_4', 'SatisfactionScore_5', 'MaritalStatus_Divorced',
       'MaritalStatus_Married', 'MaritalStatus_Single', 'Complain_0.0',
       'Complain_1.0']].copy()
y = final_processed_data['Churn']
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size = 0.8, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
norm = MinMaxScaler().fit(X_train)

# transform training data
X_train_norm = norm.transform(X_train)

# transform testing dataabs
X_test_norm = norm.transform(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=42)
rfc.fit(X_train, y_train)

print(type(rfc))
scores = cross_val_score(rfc,X_train,y_train,cv=5, scoring='roc_auc')
print ("CV score :",scores.mean())
rfc.fit(X_train, y_train)
pred = rfc.predict(X_test)
print(classification_report(y_test, pred))
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=42)
lr.fit(X_train, y_train)

print(type(lr))
scores = cross_val_score(lr,X_train,y_train,cv=5,scoring='roc_auc')
print ("CV score :",scores.mean())
lr.fit(X_train, y_train)
pred = lr.predict(X_test)
print(classification_report(y_test, pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=42, max_depth=20)
dt.fit(X_train, y_train)

print(type(dt))
scores = cross_val_score(dt,X_train,y_train,cv=5,scoring='roc_auc')
print ("CV score :",scores.mean())
dt.fit(X_train, y_train)
pred = dt.predict(X_test)
print(classification_report(y_test, pred))
```

```python
from sklearn.svm import SVC
svm = SVC(kernel="linear", C=0.5)
svm.fit(X_train, y_train)

print(type(svm))
scores = cross_val_score(svm,X_train,y_train,cv=5,scoring='roc_auc')
print ("CV score :",scores.mean())
svm.fit(X_train, y_train)
pred = svm.predict(X_test)
print(classification_report(y_test, pred))
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(alpha=1, max_iter=1000)
mlp.fit(X_train, y_train)

print(type(mlp))
scores = cross_val_score(mlp,X_train,y_train,cv=5,scoring='roc_auc')
print ("CV score :",scores.mean())
mlp.fit(X_train, y_train)
pred = mlp.predict(X_test)
print(classification_report(y_test, pred))
```
